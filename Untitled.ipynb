```python
import rs_learn as rsl
import os, glob
import pandas as pd
from skimage.filters import threshold_otsu
import numpy as np
import matplotlib.pyplot as plt
```

```python
raster_path = os.getcwd()
raster_name = 'quezon_city'    
rd = rsl.Image_to_array(raster_path = raster_path,
                        raster_name = raster_name,
                        raster_extension = 'tif')
```

```python
arr = rd.convert_to_array()
```

```python
arr
```

```python
#get threshold
thresh = threshold_otsu(arr[~np.isnan(arr)])
```

```python
arr[np.isnan(arr)] = 0
```

```python
arr2 = np.where((arr < thresh) & (arr > 0), 1, 0)

# arr2 = (np.where((a > 2) & (a < 6), -1, 100))
```

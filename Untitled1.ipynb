```python
import rs_learn as rsl
import numpy as np
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu

def rolling_window(array, window, step_size):
    shape = array.shape[:-1] + (array.shape[-1] - window + 1 - step_size + 1, window)
    strides = array.strides + (array.strides[-1] * step_size,)
    subset = np.lib.stride_tricks.as_strided(array, 
                                             shape = shape, 
                                             strides = strides)
    thresh = threshold_otsu(subset[~np.isnan(subset)])
    array_sub = np.where(subset < thresh, 1, 0)
    return array_sub
```

```python
raster_path = r'C:\Users\Dlaniger\Projects\MAPABLE\OIL_SPILL\S1\processed\processing\iloilo\rasters\roi_raster'
raster_name = 'raster-02'
rd = rsl.Image_to_array(raster_path = raster_path,
                        raster_name = raster_name,
                        raster_extension = 'tif')

array = rd.convert_to_array()
# binary = rd.binarize_array(array,
#              raster_name,
#              raster_name)

# rd.convert_to_image(binary,
#                     raster_name)
array_sub = rolling_window(array, 100, step_size = 2)
```
